### 주의! 이 때 복사본을 저장해야 한다. 작업중인 파일은 그 자체를 옮길 수 없다.

### TIP! Colab 에서 생성된 jupyter notebook 파일은 전부 default로 Colab Notebooks 디렉토리 안에 저장된다.
### 따라서, 원하는 디렉토리 안에 저장해야 할 경우, 반드시 복사본을 만들어 저장하도록 작업해야 한다.

```
import shutil

try :
  # shutil.copy(<복사할 파일>, <복사될 파일>)
  src_file = colab_directory + '/Untitled0.ipynb' # 복사할 파일
  desc_file = 'Untitled0.ipynb' # 복사될 파일
  shutil.copy(src_file, desc_file)
except Exception as err:
  print(err)
```
