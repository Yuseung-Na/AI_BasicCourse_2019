### save_dir : ./<save_dir>/"keyword"/ 
### 최소 100개 이상의 image
### keyword : dog, cat, monkey, racoon, bear, pig, lion
### parent directory : mammal_photos

```python
from icrawler.builtin import GoogleImageCrawler
import os

# input keywords
search_words = input("input keyword to download : ")

# replace "," to ""
if "," in search_words :
    search_words = search_words.replace(',', '')

search_list = search_words.split()

# debug
print(search_list)

# input max_limits
max_limits = int(input("max crawling count : "))


# input image_save dir
save_dir = input("input parent directory : ")

# default size
size_opt = '>640x480' # medium, large, icon, >[]x[], =[]x[] ([] is an integer) e.g ">640x480" (x must be small letter) 800X600

# type
# photo_type = 'face' # photo, face, clipart, linedrawing, animated



# date
# date_opt = 'pastyear' # pastday, pastweek, or a tuple of dates. e.g ((2016, 1, 1), (2017, 1, 1)) or ((2016, 1, 1), None) *None means until today

# color
# color_opt = "color" #“blackandwhite”, “red”, “orange”, “yellow”, “green”, “teal”, “blue”, “purple”, “pink”,“white”, “gray”, “black”, “brown”

# default license
# license_opt = 'commercial,modify' # noncommercial, commercial, noncommercial,modify, commercial,modify
# 개인 사용 목적일 경우 영리 이미지 사용가능
# license_opt = 'None'

# save directory (create)
curr_dir = os.getcwd()

if not os.path.exists(save_dir):
    os.mkdir(save_dir)

os.chdir(save_dir)
parent_dir = os.getcwd()

# debug
print("current directory : {}".format(parent_dir))

for img_dir in search_list:
    if not os.path.exists(img_dir) :
        os.mkdir(img_dir)
        
    os.chdir(parent_dir)
    google_crawler = GoogleImageCrawler( 
        feeder_threads = 1,
        parser_threads = 2, 
        downloader_threads = 4,
        storage={ 'root_dir' : parent_dir + "/" + img_dir })
    filters = dict(
        size = size_opt)
#         type = photo_type ) 
#         license = license_opt)
    google_crawler.session.verify = False
    google_crawler.crawl(keyword = img_dir , filters = filters, max_num = max_limits, file_idx_offset = 'auto')
    
    # debug
    print("keyword {} {} images have been saved.".format( img_dir, max_limits))
    
# restore to original directory
os.chdir(curr_dir)
```

```python
# 현재 작업 디렉토리 출력
print(os.getcwd())

try :
  img_dir = "/content/drive/My Drive/data/mammals/"
except Exception as err :
  print(err)
```

### Image 파일 jpeg 으로 통일

```python
from PIL import Image

img_ext_list = [".jpg", ".JPG", ".gif", ".GIF", ".png", ".PNG", ".jpeg", ".JPEG"]

def verify_image(img_file):
  # test image
  try:
    v_image = Image.open(img_file)
    v_image.verify()
    v_image.close()
    # if image is valid
    return True
  except OSError:
    return False

for root, dirs, files in os.walk(img_dir):
  # to add number
  cnt = 1
  for file in files:
    currentFile=os.path.join(root, file)
    
    # print("debug : ", currentFile, "\n")
    file, ext = os.path.splitext(currentFile)

    if ext in img_ext_list :
      ext = ".jpeg"
    # test image
    if verify_image(currentFile):
      new_file_name = os.path.dirname(currentFile)
      try:
        # debug dirname
        final_dir = os.path.dirname(currentFile) + "/"
        # print("saved in : ", final_dir, "\n")
        new_file_name = final_dir + os.path.basename(os.path.dirname(currentFile)) + str(cnt) + ext
        # print("finally saved in : ", new_file_name)
        os.rename(currentFile, new_file_name )
        
        # add cnt
        cnt += 1
        print("file saved: ", new_file_name , "\n")
        
      except Exception as err:
        print(err)
    # else :
    #   print("\n\n[Illegal file ", currentFile , " removed\n\n")
    #   os.remove(currentFile)
```

### 디렉토리 목록 list로 적용

```python
dir_list = os.listdir(img_dir)

final_dir_list = []

for dirname in dir_list:
  final_dir_list.append(img_dir + dirname)
  # print(img_dir + dirname)

for dirname in final_dir_list:
  print(dirname)
```

### 유효 이미지 검사하여 유효 이미지만 저장

```python
import tensorflow as tf
import glob

for dirname in final_dir_list:
  for i, image_name in enumerate(glob.glob(os.path.join(dirname, '*.jpeg'))):

    print(i, image_name)

    with tf.Graph().as_default():

        image_contents = tf.read_file(image_name)

        image = tf.image.decode_jpeg(image_contents, channels=3)

        # init_op = tf.initialize_all_tables()

        with tf.Session() as sess:

            # sess.run(init_op)
            try:
              sess.run(tf.global_variables_initializer())
              tmp = sess.run(image)

            except Exception as err:
              # print(err)
              print("[illegal file : ", image_name, " will be removed.\n")
              os.remove(image_name)
```
